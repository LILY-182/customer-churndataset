```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Churn.csv')
df
```

```python
df.drop('customerID', axis=1, inplace=True)
df
```

```python
df.groupby('Dependents').size()
df['Dependents'] = df['Dependents'].replace('No','0')
df['Dependents'] = df['Dependents'].replace('Yes','1')
df
```

```python
df.groupby('gender').size()
df['gender'] = df['gender'].replace('Male','0')
df['gender'] = df['gender'].replace('Female','1')
df
```

```python
#3. Multiple lines
df.groupby('PhoneService').size()
df['PhoneService'] = df['PhoneService'].replace('No',0)
df['PhoneService'] = df['PhoneService'].replace('Yes',1)
df
```

```python
df.groupby('MultipleLines').size()
df['MultipleLines'] = df['MultipleLines'].replace('No',0)
df['MultipleLines'] = df['MultipleLines'].replace('Yes',1)
df['MultipleLines'] = df['MultipleLines'].replace('No Phone',2)
df
```

```python
df.groupby(['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']).size()
```

```python
column_to_encode = ['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']
for column in column_to_encode:
  df.groupby(column).size()
  df[column] = df[column].replace('No',0)
  df[column] = df[column].replace('Yes',1)
# 0r df['column'] = df['column'].replace({
   # 1:'YES',
   # 0:'NO'
#})
df
```

```python

df
```

```python
df.groupby('InternetService').size()
df['InternetService'] = df['InternetService'].replace('DSL',2)
df['InternetService'] = df['InternetService'].replace('Fiber',1)
df['InternetService'] = df['InternetService'].replace('No',0)

df.groupby('InternetService').size()
```

```python
#payment method
df.groupby('PaymentMethod').size()
df['PaymentMethod'] = df['PaymentMethod'].replace({
    'Bank': 1,
    'Credit card': 0,
    'Electronic': 2,
    'Mailed': 3
})

df
```

```python
df.groupby('Contract').size()
df['Contract'] = df['Contract'].replace({
    'Month': 0,
    'One year': 1,
    'Two year': 2
})
df
```
